```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance_data.csv')
df
```

```python
plt.xlabel('age')
plt.scatter(df.age, df.bought_insurance, color = 'red', marker = '+')
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['age']], df.bought_insurance)
```

```python
plt.xlabel('age')
plt.scatter(df.age, df.bought_insurance, color = 'red', marker = '+')
plt.plot(df.age, reg.predict(df[['age']]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size = 0.9)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.predict(X_train)  # all the x_train values who will buy insurance and those who wont
```

```python
X_train
```

```python
reg.predict(X_test)  # since there were only 3 values in X_test so output is for those 3 values
```

```python
reg.score(X_test, y_test)
```
